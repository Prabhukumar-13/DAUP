```python
data.replace('male',1,inplace=True)
data.replace('female',0,inplace=True)
data.replace('yes',1,inplace=True)
data.replace('no',0,inplace=True)
data.replace('northeast',0,inplace=True)
data.replace('northwest',1,inplace=True)
data.replace('southeast',2,inplace=True)
data.replace('southwest',3,inplace=True)
```

# **Part-3: Filter data Smokers Between 20 and 40**

```python

smokers_data = data[data['smoker'] == 1]
age_data = data[(data['age'] >= 20) & (data['age'] <= 40)]

print("Smokers Data:")
print(smokers_data)

print("\nAge Data (20-40):")
print(age_data)
plt.figure(figsize=(10, 6))
plt.scatter(age_data['age'], age_data['smoker'], c='blue', marker='*')
plt.xlabel('Age')
plt.ylabel('Smoker (1 for yes, 0 for no)')
plt.title('Scatter Plot of Age vs Smoker Status')
plt.grid(True)
plt.show()

age_count = len(age_data)
print(f"Number of people in the specified age range: {age_count}")
```
